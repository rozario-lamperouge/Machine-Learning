# **K Nearest Neighbors**

KNN algorithms decide a number k which is the nearest Neighbor to that data point which is to be classified. If the value of k is 3 it will look for 3 nearest Neighbors to that data point. In this example, k=6. KNN finds out 6 nearest Neighbors.

image.png

## **Importing Libraries**

```
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report
```

## **Reading the data**

```
df = pd.read_csv('Machine-Learning-Course/wo_men1.csv')
df.tail(60)
```

```
df.dropna(inplace =True)
df.head(60)
```

### **Allocating the data**

```
df = np.array(df)

X = df[:,2:]
Y = df[:,1]
```

```
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size=0.2,random_state=7)
```

## **Fitting data to the model**

```
clf = KNeighborsClassifier()
clf.fit(X_train,Y_train)
```

```
score = clf.score(X_train,Y_train)
print(clf.predict([[175,44]]))
```

```
predict = clf.predict(X_test)
print(classification_report(Y_test, predict))
```

# **Support Vector Machines**

A support vector machine (SVM) is a supervised machine learning model that uses classification algorithms for two-group classification problems.

image.png

image.png

```
from sklearn.svm import SVC
```

image.png

```
clf = SVC(kernel="linear",C=1)
clf.fit(X_train,Y_train)
```

```
score = clf.score(X_train,Y_train)
print(clf.predict([[175,44]]))
```

```
predict = clf.predict(X_test)
print(classification_report(Y_test, predict))
```
